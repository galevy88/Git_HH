```python
import pandas as pd
import matplotlib.pyplot as plt

cell = str(9412)

# Read the dataframe
df = pd.read_csv(f'{cell}/sample_1_removed.csv', header=None)
df = df.drop(df.index[0])

df.columns = ['Time', 'Temp', 'Value']

# Convert 'Time' and 'Value' to numeric types
df['Time'] = pd.to_numeric(df['Time'])
df['Value'] = pd.to_numeric(df['Value'])

# Plot df with smaller markers
plt.plot(df['Time'], df['Value'], 'x', label='Data', markersize=1)

# Add title and labels
plt.title('Dataset')
plt.xlabel('Time (ms)')
plt.ylabel('Conductivity (G)')

# Adjust the legend position outside the plot and to the right side
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```
